```python
#to change the integer values to categorical values
#Method 2: using map and astype

#season to categorical values
bike_day['seasons'] = bike_day.season.map({1 : "Winter", 2 : "Spring", 3 : "Summer", 4 : "Fall"}).astype('category')
bike_hour['seasons'] = bike_hour.season.map({1 : "Winter", 2 : "Spring", 3 : "Summer", 4 : "Fall"}).astype('category')

#holiday to categorical values
bike_day['is_holiday'] = bike_day.holiday.map({1 : "yes", 0 : "no"}).astype('category')
bike_hour['is_holiday'] = bike_hour.holiday.map({1 : "yes", 0 : "no"}).astype('category')

#weekday
bike_day['weekdays'] = bike_day.weekday.map({0 : "Sunday", 1 : "Monday", 2 : "Tuesday", 3 : "Wednesday", 4 : "Thursday", 5 : "Friday" , 6 : "Satarday"}).astype('category')
bike_hour['weekdays'] = bike_hour.weekday.map({0 : "Sunday", 1 : "Monday", 2 : "Tuesday", 3 : "Wednesday", 4 : "Thursday", 5 : "Friday" , 6 : "Satarday"}).astype('category')

#workingday to categorical values
bike_day['is_workingday'] = bike_day.workingday.map({1 : "yes", 0 : "no"}).astype('category')
bike_hour['is_workingday'] = bike_hour.workingday.map({1 : "yes", 0 : "no"}).astype('category')

#weathersit to categorical values
bike_day['weather'] = bike_day.weathersit.map({1 : "Sunny", 2 : "Cloudy" , 3 : "Rainy", 4 :"Heavy rain"}).astype('category')
bike_hour['weather'] = bike_hour.weathersit.map({1 : "Sunny", 2 : "Cloudy" , 3 : "Rainy", 4 :"Heavy rain"}).astype('category')

#delete orignal columns
bike_day = bike_day.drop(columns=['season', 'holiday', 'workingday', 'weathersit'])
bike_hour= bike_hour.drop(columns=['season', 'holiday', 'workingday', 'weathersit'])

```

```python
#rename dteday to date
bike_day = bike_day.rename(columns={'dteday' : 'date'})
bike_hour = bike_hour.rename(columns={'dteday' : 'date'})
```

```python
#to denormalize temp, atemp, hum, and windspeed features 

#temp
bike_hour['temp'] = bike_hour['temp']* (+39 - (-8)) + (-8)
bike_day['temp'] = bike_day['temp']* (+39 - (-8)) + (-8)

#atemp
bike_hour['atemp'] = bike_hour['atemp']* (+50 - (-16)) + (-16)
bike_day['atemp'] = bike_day['atemp']* (+50 - (-16)) + (-16)

#hum
bike_hour['hum'] = bike_hour['hum']* 100
bike_day['hum'] = bike_day['hum']* 100

#windspeed
bike_hour['windspeed'] = bike_hour['windspeed']* 67
bike_day['windspeed'] = bike_day['windspeed']* 67


bike_hour.head(10)
bike_hour[bike_hour['mnth'] == 6]

```

```python
#adding new features 


#bike_day
bike_day['hot_day'] = bike_day['temp'].apply(lambda value: 'yes' if value >= 36 else 'no' ) 
bike_day['windy_day'] = bike_day['windspeed'].apply(lambda value: 'yes' if value >= 30 else 'no' ) 
bike_day['high_hum'] = bike_day['hum'].apply(lambda value: 'yes' if value >= 60 else 'no' ) 

#bike hour
bike_hour['hot_day'] = bike_hour['temp'].apply(lambda value: 'yes' if value >= 36 else 'no' ) 
bike_hour['windy_day'] = bike_hour['windspeed'].apply(lambda value: 'yes' if value >= 30 else 'no' ) 
bike_hour['high_hum'] = bike_hour['hum'].apply(lambda value: 'yes' if value >= 60 else 'no' ) 

#bike_hour.head(40)
seasons = bike_hour.groupby('seasons')['mnth'].unique().apply(list)
seasons


```

2. EDA 

Time based:

```python
#Hourly trend: most of the businesses have rush hours and weak hours. It should not be an exception for bike rentals as well.

#Group the data by hour and calculate the average count of bike rentals:
hourly_counts = bike_hour.groupby('hr').agg({'cnt':'mean'}).reset_index()

#Visualize the hourly trends using a line plot:
sns.lineplot(x='hr', y='cnt', data=hourly_counts)
plt.xlabel('Hour')
plt.ylabel('Average Count of Bike Rentals')
plt.title('Hourly Trends of Bike Rentals')
plt.grid()
plt.show()
```

```python
#time category

# Define the time categories
time_bins = [0, 6, 12, 17, 20, 24]
time_labels = ['Night', 'Morning', 'Noon', 'Afternoon', 'Evening']

# Create a new column 'time_category' based on the 'hr' column
bike_hour['time_category'] = pd.cut(bike_hour['hr'], bins=time_bins, labels=time_labels, include_lowest=True)

# Group the data by time_category and calculate the average count of bike rentals for each category
hourly_counts = bike_hour.groupby('time_category').agg({'cnt':'mean'}).reset_index()

# Create a bar plot of the hourly trends of bike rentals by time category
sns.barplot(x='time_category', y='cnt', data=hourly_counts, order=time_labels)
plt.xlabel('Time Category')
plt.ylabel('Average Count of Bike Rentals')
plt.title('Hourly Trends of Bike Rentals by Time Category')
plt.show()

```

```python

#Group the data by day and calculate the average count of bike rentals:
daily_counts = bike_hour.groupby('weekdays').agg({'cnt':'mean'}).reset_index()

#Visualize the daily trends using a line plot:
sns.lineplot(x='weekdays', y='cnt', data=daily_counts)
plt.xlabel('Week Days')
plt.ylabel('Average Count of Bike Rentals')
plt.title('Daily Trends of Bike Rentals')
plt.grid()
plt.show()
```

```python

# Group the data by month and calculate the average count of bike rentals for each month
monthly_counts = bike_hour.groupby(pd.Grouper(key='mnth')).agg({'cnt':'mean'}).reset_index()

# Set the figure size and font sizes
plt.rcParams["figure.figsize"] = (10, 6)
plt.rcParams["font.size"] = 14

# Create a line plot of the monthly trends of bike rentals
sns.lineplot(x='mnth', y='cnt', data=monthly_counts)
plt.xlabel('Month')
plt.ylabel('Average Count of Bike Rentals')
plt.title('Monthly Trends of Bike Rentals')
plt.grid()

# Show the plot
plt.show()
```

```python

# Group the data by month and calculate the total count of bike rentals for each month
monthly_counts = bike_day.groupby('mnth')['cnt'].sum().reset_index()

# Set the figure size and font sizes
plt.rcParams["figure.figsize"] = (10, 6)
plt.rcParams["font.size"] = 14

# Create a bar chart of the monthly distribution of bike rentals
sns.barplot(x='mnth', y='cnt', data=monthly_counts)
plt.xlabel('Month')
plt.ylabel('Total Count of Bike Rentals')
plt.title('Monthly Distribution of Bike Rentals')
plt.grid()

# Show the plot
plt.show()
```

2. Daily Trend: weekdays vs weekends and registered users vs casual users.

```python
daily_counts = bike_day.groupby(['weekdays', 'is_workingday']).mean().reset_index()

weekday_counts = daily_counts[daily_counts['is_workingday'] == 'yes']
weekend_counts = daily_counts[daily_counts['is_workingday'] == 'no']

registered_weekday_avg = weekday_counts['registered'].mean()
casual_weekday_avg = weekday_counts['casual'].mean()

registered_weekend_avg = weekend_counts['registered'].mean()
casual_weekend_avg = weekend_counts['casual'].mean()
```

```python
sns.set_style('whitegrid')

data = [['Weekday', registered_weekday_avg, casual_weekday_avg],
        ['Weekend', registered_weekend_avg, casual_weekend_avg]]
plot_data = pd.DataFrame(data, columns=['Day Type', 'Registered', 'Casual'])

sns.barplot(x='Day Type', y='value', hue='variable', data=pd.melt(plot_data, ['Day Type']))
plt.xlabel('')
plt.ylabel('Average Count of Bike Rentals')
plt.title('Daily Trends of Bike Rentals')
plt.show()
```

Weather based:

3. Rain: the demand for bikes might be changed on a rainy day as compared to a sunny day. Similarly, people prefer to go out on less humid days

```python
# Calculate the average number of bike rentals on rainy days
rainy_avg = bike_day.loc[bike_day.weather == 'rainy', 'cnt'].mean()

# Calculate the average number of bike rentals on sunny days
sunny_avg = bike_day.loc[bike_day.weather == 'sunny', 'cnt'].mean()

#calculate the average number of bike rentals on cloudy days
cloudy_avg = bike_day.loc[bike_day.weather == 'cloudy', 'cnt'].mean()

#calculate the average number of bike rentals on days with heavy rain
heavy_rain_avg = bike_day.loc[bike_day.weather == 'heavy_rain', 'cnt'].mean()


print("Average number of bike rentals on rainy days:", rainy_avg)
print("Average number of bike rentals on sunny days:", sunny_avg)
print("Average number of bike rentals on cloudy days:", cloudy_avg)
print("Average number of bike rentals on heavy_rainy days:", heavy_rain_avg)
```

```python
#create plot to visualize average number of bike rentals on different weather conditions
sns.boxplot(x='weather', y='cnt', data=bike_day)
```

```python
#using bar chart 
weather_groups = bike_day.groupby('weather').mean()
plt.bar(weather_groups.index, weather_groups.cnt)
```

```python
# Calculate the average number of bike rentals on high-humidity days
high_hum_avg = bike_day.loc[bike_day.high_hum == 'yes', 'cnt'].mean()

# Calculate the average number of bike rentals on low-humidity days
low_hum_avg = bike_day.loc[bike_day.high_hum == 'no', 'cnt'].mean()

print("Average number of bike rentals on high-humidity days:", high_hum_avg)
print("Average number of bike rentals on low-humidity days:", low_hum_avg)
```

```python
# Create a bar plot of the average number of bike rentals on high-humidity and low-humidity days
plt.bar(['High Humidity', 'Low Humidity'], [high_hum_avg, low_hum_avg])
plt.ylabel('Average Number of Bike Rentals')
plt.title('Bike Rentals by Humidity Level')
plt.show()
```

```python
#total number of rentals by user type 

# Calculate the total count of registered and casual users
registered_count = bike_hour['registered'].sum()
casual_count = bike_hour['casual'].sum()
total_count = registered_count + casual_count

# Calculate the percentage of registered and casual users
registered_percent = (registered_count / total_count) * 100
casual_percent = (casual_count / total_count) * 100

# Create a pie chart showing the percentage of registered and casual users
labels = ['Registered Users', 'Casual Users']
sizes = [registered_percent, casual_percent]
colors = ['lightblue', 'lightgreen']
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=90)
plt.axis('equal')
plt.title('Percentage of Registered and Casual Users')
plt.show()
```

We can analyze the usage patterns of registered users and casual users separately. For example, we could look at the days of the week or times of day when each group tends to rent bikes more frequently.

```python
# Group the data by user type and weekdays and calculate the average count of bike rentals for each group
user_counts_by_weekday = bike_hour.groupby(['casual','registered', 'weekdays']).agg({'cnt':'mean'}).reset_index()

# Create a grouped bar chart of the average count of bike rentals by user type and weekdays
sns.catplot(x='weekdays', y='cnt', data=user_counts_by_weekday, kind='bar')
plt.xlabel('Weekdays')
plt.ylabel('Average Count of Bike Rentals')
plt.title('Average Count of Bike Rentals by User Type and Weekdays')
plt.show()

# Group the data by user type and time_category and calculate the average count of bike rentals for each group
user_counts_by_time = bike_hour.groupby(['casual','registered', 'time_category']).agg({'cnt':'mean'}).reset_index()

# Create a grouped bar chart of the average count of bike rentals by user type and time category
sns.catplot(x='time_category', y='cnt', data=user_counts_by_time, kind='bar')
plt.xlabel('Time Category')
plt.ylabel('Average Count of Bike Rentals')
plt.title('Average Count of Bike Rentals by User Type and Time Category')
plt.show()
```

```python
#season wise hourly distribution of bike rentals 

# Create a pivot table to get season-wise hourly distribution of rental bikes
season_hourly = bike_hour.pivot_table(index='hr', columns='seasons', values='cnt', aggfunc='mean')

# Plot the pivot table using a line plot
season_hourly.plot(kind='line', figsize=(10,6))

# Add labels and title to the plot
plt.title('Season-wise Hourly Distribution of Rental Bikes')
plt.xlabel('Hour of the Day')
plt.ylabel('Average Number of Rental Bikes')
plt.legend(title='Season', loc='upper left')
plt.xticks(np.arange(0, 24, 1))
plt.show()
```

```python
#holiday wise hourly distribution of bike rentals 

# Create a pivot table to get holiday-wise hourly distribution of rental bikes
holiday_hourly = bike_hour.pivot_table(index='hr', columns='is_holiday', values='cnt', aggfunc='mean')

# Plot the pivot table using a line plot
holiday_hourly.plot(kind='line', figsize=(10,6))

# Add labels and title to the plot
plt.title('Holiday-wise Hourly Distribution of Rental Bikes')
plt.xlabel('Hour of the Day')
plt.ylabel('Average Number of Rental Bikes')
plt.legend(title='Holiday', loc='upper left')
plt.xticks(np.arange(0, 24, 1))
plt.show()

```
